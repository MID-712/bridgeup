```python
def date_adder(t1):
    start_date = datetime.datetime(1950, 1, 1)
    delta_t = datetime.timedelta(days= t1)
    days_now = start_date + delta_t
    return days_now
```

```python
def date_adder_full(t1):
    start_date = datetime.datetime(1950, 1, 1)
    delta_t = datetime.timedelta(days= t1)
    days_now = start_date + delta_t
    return (days_now.strftime("%-m /%d /%Y") )
```
